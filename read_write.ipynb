```python
df.to_excel('file_name.xlsx') >> this will create excel file from dataframe 
```

```python
df = pd.read_csv('movies_data.csv')
df
```

```python
df.to_excel('movies_name.xlsx')
```

```python
df_action = df.groupby('genre').get_group('Action')
df_action.to_excel('action_movies.xlsx')
```

```python
df_crime = df.groupby('genre').get_group('Crime')
df_crime.to_excel('crime_movies.xlsx',sheet_name='Crime_Movies')
```

```python
### Create a excel with multiple sheet

df_action.to_excel('movies.xlsx',sheet_name='Action_Movies')
df_crime.to_excel('movies.xlsx',sheet_name='Crime_Movies')
```

## ExcelWriter

```python
writer = pd.ExcelWriter('mov.xlsx',engine='xlsxwriter')

df_action.to_excel('movies.xlsx',sheet_name='Action_Movies')
df_crime.to_excel('movies.xlsx',sheet_name='Crime_Movies')

writer.save()
```

```python
writer = pd.ExcelWriter('mov.xlsx',engine='xlsxwriter')

df_action.to_excel(writer,sheet_name='Action_Movies')
df_crime.to_excel(writer,sheet_name='Crime_Movies')

writer.save()
```

```python
writer = pd.ExcelWriter('mov1.xlsx',engine='xlsxwriter')

df_action.to_excel(writer,sheet_name='Action_Movies',index=False)
df_crime.to_excel(writer,sheet_name='Crime_Movies',index=False)

writer.save()
```
